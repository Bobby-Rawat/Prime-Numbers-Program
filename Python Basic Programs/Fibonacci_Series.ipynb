<h1>What is the Fibonacci Series?</h1>
<p><h5>The Fibonacci series is the sequence where each number is the sum of the previous two numbers of the sequence. The first two numbers of the Fibonacci series are 0 and 1 and are used to generate the Fibonacci series.</h5><p>

The Fibonacci sequence was conceptualized by Leonardo of Pisa (known as Fibonacci) in the early 13th century. He introduced it in his book "Liber Abaci" while solving a problem on rabbit breeding, leading to the sequence where each number is the sum of the two preceding ones (starting with 0 and 1).

Its widespread usage today is primarily due to its intriguing properties, occurrence in natural phenomena, and applications across various disciplines, making it a fundamental sequence in mathematics, computer science, nature, and beyond.

<h1>Fibonacci Series - Code</h1>

```python
n = int(input("Enter Length of Fibonacci Sequence : "))

n1 = 0
n2 = 1
count = 2

if n < 0 :
  print("Enter Only Positives Values")
elif n == 1 :
  print(n1,end=" ")
else :
     print(n1,end=" ")
     print(n2,end=" ")
     while count < n:
      n3 = n1+n2
      print(n3,end=" ")
      count+=1
      n1 = n2
      n2 = n3
```
