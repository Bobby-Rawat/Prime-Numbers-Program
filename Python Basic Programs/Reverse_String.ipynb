<h1> Reverse String</h1>
<p> This code interacts with the user, asking for a string input. Upon receiving the input, it utilizes a special method called "slicing" in Python, denoted by [::-1]. This slicing technique reverses the arrangement of characters within the string, effectively creating a backward version of the original input. Finally, the code presents this reversed string back to the user for display or potential utilization.</p>

<h3> [  :  :  -1] is a Python string slicing technique used to reverse a string.</h3>
Here's a simple breakdown:

<p> The first colon : indicates the start of the slicing. </p>
<p> The second colon : separates the start and end indices (in this case, absent means "all characters"). </p>
<p> The -1 specifies the step size. A negative step size (-1) indicates a reverse traversal. </p>

```python
string_input = input("Enter a Value : ")
reversed_string =  string_input[::-1]
print(reversed_string)
```
