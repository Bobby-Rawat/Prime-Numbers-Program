<h2> What is Palindrome Program in Python? </h2>

A Palindrome can be defined as a string or a number that displays the same value when its position is reversed. When the reverse word and the actual word, (a number or a string) are compared, they produce the same result.

```python
value = input("Enter the Value : ")
reversed_value = value[::-1]

if value ==  reversed_value:
  print("It is a Palindrome")
else:
  print("It's not a Palindrome")
```

<h2>Conclusion</h2>

Palindrome is a value when compared from the backward direction, and the forward direction comes to be the same. It is used in various mathematical puzzles.
