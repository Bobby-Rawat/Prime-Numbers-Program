<h2> Even / Odd Program </h2>

```python
num = int(input("Enter a number:"))
if (num % 2) == 0:
    print("{0} is Even".format(num))
else:
    print("{0} is Odd".format(num))
```

<h2> Loop Method - Even / Odd</h2>

```python
def yes_no(message):
    while True:
        user_input = input(message).lower()
        if user_input == "yes":
            return True
        elif user_input == "no":
            return False
        else:
            print("Please, use 'yes' or 'no'.")

def get_number():
    while True:
        user_input = input("Enter a number: ")
        if user_input.isdigit():  # Check if the input consists of digits
            return int(user_input)
        else:
            print("Please enter a valid number.")

def main():
    while True:
        number = get_number()

        if number % 2 == 1:
            print(f"{number} is odd.")
        else:
            print(f"{number} is even.")

        if not yes_no("Do you want to try again?\n"):
            print("Goodbye!")
            break

if __name__ == "__main__":
    main()
```
